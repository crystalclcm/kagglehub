T1 - Section 3: EDA

```python
# Group by date_served and count number of unique user_id's
daily_users = marketing.groupby(['date_served'])['user_id'].nunique()

# Print head of daily_users
print(daily_users)
```

```python
# import python matplotlib
import matplotlib.pyplot as plt

# Plot daily_subscribers
daily_users.plot()

# Include a title and axis labels
plt.title('Daily users')
plt.xlabel('Date')
plt.ylabel('Number of users')

# Rotate the x-axis labels by 45 degrees
plt.xticks(rotation = 45)

# Display the plot
plt.show()

```

```python
import numpy as np
marketing['is_house_ads']=np.where(marketing['marketing_channel']=='House Ads', True, False)
```

```python
house_ads = marketing[marketing['is_house_ads'] == True]
print(house_ads)
```

```python
total_house_ads = house_ads['user_id'].nunique()
print(total_house_ads)

subscribers_house_ads = house_ads[house_ads['converted'] == True]\
                ['user_id'].nunique()
print(subscribers_house_ads)

retainers_house_ads = house_ads[house_ads['is_retained'] == True]\
                [house_ads['converted'] == True]['user_id'].nunique()
print(retainers_house_ads)

conv_rate_house_ads = subscribers_house_ads/total_house_ads
print(conv_rate_house_ads )

ret_rate_house_ads = retainers_house_ads/subscribers_house_ads
print(ret_rate_house_ads)
```

```python
total_channels= marketing.groupby(['marketing_channel'])\
                ['user_id'].nunique()
print(total_channels)

subscribers_channels = marketing[marketing['converted'] == True]\
              .groupby(['marketing_channel'])\
                ['user_id'].nunique()
print(subscribers_channels)

retainers_channels = marketing[marketing['is_retained'] == True]\
                [marketing['converted'] == True]\
                .groupby(['marketing_channel'])\
                ['user_id'].nunique()
print(retainers_channels)

conv_rate_channels = subscribers_channels/total_channels
print(conv_rate_channels)

ret_rate_channels = retainers_channels/subscribers_channels
print(ret_rate_channels)
```
